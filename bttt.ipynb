# Bài tập tính toán
- Vẽ các tín hiệu sau và tìm ý nghĩa vật lý của từng phép toán.
- x3(n) = x(n).rect3(n-1) 
- x4(n) = x(2n)
- x5(n) = x(2n-1)
- x6n = 0.5x(n) + 0.5x(-n)
- x7(n) = 0.5x(n) – 0.5x(-n)
- x8(n) = 0.5x(n) + 0.5(-1)^n.x(n)
- x9(n) = 0.5x(n) + 0.5.(-1)^n.x(n)

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
n = np.arange(-15,16) #n
#x(n)
x_n = np.zeros_like(n)
x_n[(n>=-3) & ( n <= 4)] = [2,2,2,0,1,2,3,4]
#rect3(n)
rect3_n = np.zeros_like(n)
rect3_n[(n>=0) & ( n <=2 )] =1
rect3_n_1= np.zeros_like(n)
rect3_n_1[(n>=1) & ( n <= 3)] =1
#x3
x3  = x_n*rect3_n_1
#x4
x4 = []
for k in n:
    idx = 2*k
    if idx in n:
        pos = np.where(n == idx)[0][0]
        x4.append(x_n[pos])
    else:
        x4.append(0)
x4 = np.array(x4)
#x5
x5 = []
for k in n:
    idx = 2*k - 1
    if idx in n:
        pos = np.where(n == idx)[0][0]
        x5.append(x_n[pos])
    else:
        x5.append(0)
x5 = np.array(x5)
#x6
x__n = []
for k in n:
    idx = -k
    if idx in n:
        pos = np.where(n == idx)[0][0]
        x__n.append(x_n[pos])
    else:
        x__n.append(0)
x__n = np.array(x__n)
x6 = 0.5 * x_n + 0.5 * x__n
#x7
x7 = 0.5*x_n - 0.5*x__n
#x8
x8 = 0.5*x_n + 0.5*np.power(-1.0,n)*x_n



```

```python
plt.figure(figsize=(12, 4))
plt.vlines(n, 0, x_n, colors='blue')
plt.scatter(n, x_n, color='red')
plt.title('x(n)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.vlines(n, 0, x3, colors='blue')
plt.scatter(n, x3, color='red')
plt.title('x3(n)')
plt.xlabel('n')
plt.ylabel('x3(n)')
plt.grid(True)
plt.show()
```
