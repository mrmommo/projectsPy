# Lập trình mạch tạo dao động hình sin:
 - Tác động vào xung đơn vị và hiển thị 100 mẫu đầu tiên của y(n): 
   y(n) – 1.41.y(n-1) + y(n-2) = x(n-1)
- Hãy xác định tần số của tín hiệu hình sin thu được.
- Từ đó hãy tìm các hệ số của phương trình truy hồi bậc hai để tạo mạch dao động hình sin với tần số w0 bất kỳ. 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# 100 mẫu
N = 100
# khởi tạo x(n) = delta(n)
x = np.zeros(N)
x[0] = 1
# khoi tao y(n)
y = np.zeros(N)
for n in range(1,N):
    y[n] = x[n-1] + 1.41*y[n-1] - y[n-2]
    
```

```python
# Vẽ 100 mẫu đầu tiên
plt.figure(figsize=(15,4))
plt.stem(range(N), y)
plt.title("100 mẫu đầu tiên của y(n)")
plt.xlabel("n")
plt.ylabel("y(n)")
plt.grid(True)
plt.show()
```
